```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x= np.linspace(4,15,100)
y= np.sin(x)
```

```python
x
```

```python
y
```

```python
plt.figure(figsize=(5,3))
plt.plot(x,y, label="sin wave",linestyle="-",linewidth=2,color="blue")
plt.xlabel("x value")
plt.ylabel("y value")
plt.title("sin wave")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(5,3))
plt.plot(x,y, label="sin wave",linestyle="--",linewidth=2,color="blue")
plt.xlabel("x value")
plt.ylabel("y value")
plt.title("sin wave")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(5,3))
plt.plot(x,y, label="sin wave",linestyle="dotted",linewidth=2,color="blue")
plt.xlabel("x value")
plt.ylabel("y value")
plt.title("sin wave")
plt.legend()
plt.grid(True)
plt.show()
```

```python
dice = np.array([1,2,3,4,5,6]) #dice
sample  = np.random.choice(dice,4)#rolling dice 4 time in a row and getting values
print("Output of 4 samples", sample)
print("Mean of above 4 samples ",sample.mean())
```

```python
samples_1000 = [np.random.choice(dice,4).mean() for i in range(1000)]
samples_1000
```

```python
plt.figure(figsize=(5,3))
plt.hist(samples_1000, alpha=0.7, edgecolor="black",bins=100, histtype="barstacked")
plt.show()
```

```python
angles = np. linspace(0,2*np.pi,100)
radii = np.random.rand(100)
```

```python
plt.figure(figsize=(5,4))
ax = plt.subplot(111, polar=True)
ax.plot(angles, radii, color="lightblue", linewidth=0.2)
ax.fill(angles, radii, color="blue", alpha=0.3)
plt.title("polar plot")
plt.show()
```

```python
overs = np.arange(1,21)
ind = np.random.randint(1,24,20)
nz = np.random.randint(1,24,20)
```

```python
ind_total=[0]
for i in ind:
    ind_total.append(i+(ind_total[-1]))
nz_total=[0]
for i in ind:
    nz_total.append(i+(ind_total[-1]))
```

```python
plt.figure(figsize=(10,4))
plt.subplot(121)
plt.plot(overs,nz_total[1:],label="India")
plt.plot(overs, ind_total[1:], label="New Zealand")
plt.legend()

plt.subplot(122)
plt.bar(overs,nz,label="India")
plt.bar(overs, ind, label="New Zealand")
plt.legend()
plt.show()
```
