```python
import numpy as np
import pandas as pd
```

```python
arr = np.random.randint(40,99,24).reshape(6,4)
arr
```

```python
df=pd.DataFrame(arr,columns=["math","english","physics","sports"])
df["Std_Name"]=["sachin","rohit","shubhman","virat","ravindra","rinku"]
cols=["Std_Name","math","english","physics","sports"]
df=df[cols]
df.head()
```

```python
df1 = df[["Std_Name","math","english"]]
df2 = df[["Std_Name","physics","sports"]]
```

```python
df1.head(2),df2.head(2) #this made the output tuple
```

```python
type((df1.head(2),df2.head(2)))
```

```python
df1 =df1[df1["Std_Name"]!="sachin"]
df1
```

```python
df2 =df2[df2["Std_Name"]!="rinku"]
df2
```

```python
df1.merge(df2, on="Std_Name" , how = "outer")
```

```python
df1.merge(df2, on="Std_Name" , how = "inner")
```

```python
df1.merge(df2, on="Std_Name" , how = "left")
```

```python
df1.merge(df2, on="Std_Name" , how = "right")
```

```python
df1.merge(df2, how="cross")
```
